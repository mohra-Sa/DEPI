## Drop Duplicates

```python
before = len(df)
df.drop_duplicates(inplace = True)
after = len(df)
```

```python
print(before)
print()
print(after)
```

## Replace Null Values

```python
df["Parent_Education_Level"].isnull().sum()
```

```python
df = df[df["Parent_Education_Level"].notna()]
df
```

```python
df.head(10)
```

```python
df = df.reset_index(drop = True)
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.rename(columns = {"Department" : "Subject"} , inplace = True)
```

## Create table Date

```python
rng = pd.date_range("2024-09-01", "2025-06-30", periods = len(df))
df["date"] = rng
```

```python
df.head()
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df["date"] = pd.to_datetime(df["date"], errors = "coerce")
```

## Count the missing data

```python
missing_dates = df["date"].isna().sum()
print("The Count of missing dates", missing_dates)
```

```python
df.head(10)
```

```python
df["date"] = pd.to_datetime(df["date"], errors = "coerce").dt.strftime("%Y-%m-%d")
```

```python
df.head()
```

```python
df.duplicated().sum()
```

## Create The Performance Column

```python
df["performance"] = pd.cut(df["Total_Score"], bins = [0, 50, 75, 100], labels = ["Low", "Medium", "High"], include_lowest = True)
```

```python
final_df = df[["Student_ID", "First_Name", "Last_Name", "Subject", "Attendance (%)", "date", "performance"]]
```

```python
final_df.to_csv("Students_clean.csv", index = False)
```
